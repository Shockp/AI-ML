#### **Mini-Project: Visualizing Vector Operations**

##### **Objective**
Create a Python program that:
1. Takes two vectors as input from the user.
2. Performs operations such as:
   - Addition ($\mathbf{u} + \mathbf{v}$)
   - Subtraction ($\mathbf{u} - \mathbf{v}$)
   - Scalar multiplication ($c\mathbf{u}$)
   - Dot product ($\mathbf{u} \cdot \mathbf{v}$)
3. Visualizes the original vectors and the results of these operations on a 2D plane.
4. Dynamically adjusts the axis scaling based on the input vectors.
5. Saves the final visualization as an image file for future reference.

##### **Steps to Build**
1. **Input Collection**:
   - Accept vector inputs $\mathbf{u}$ and $\mathbf{v}$.
   - Accept a scalar value for multiplication.

2. **Perform Operations**:
   - Addition: $\mathbf{u} + \mathbf{v}$
   - Subtraction: $\mathbf{u} - \mathbf{v}$
   - Scalar Multiplication: $c\mathbf{u}$
   - Dot Product: $\mathbf{u} \cdot \mathbf{v}$

3. **Visualization**:
   - Plot the original vectors $\mathbf{u}$ and $\mathbf{v}$.
   - Plot their sum $\mathbf{u} + \mathbf{v}$, difference $\mathbf{u} - \mathbf{v}$, and scalar multiplication $c\mathbf{u}$.
   - Label each vector and ensure the axis scales dynamically.
   - Annotate the origin and other key points for clarity.

4. **Save the Plot**:
   - Export the plot as an image file (e.g., `vector_visualization.png`).

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Input collection
def input_vector2D():
    x = int(input("Enter the value of the x: "))
    y = int(input("Enter the value of the y: "))
    return [x, y]

u = np.array(input_vector2D())
v = np.array(input_vector2D())

print(f"Vector u = {u}")
print(f"Vector v = {v}")

# 2. Perform operations
scalar_constant = int(input("Enter the number for the scalar multiplication operation with the vector u: "))

addition = u + v
subtraction = u - v
scalar_multiplication = scalar_constant * u
dot_product = np.dot(u, v)

print(f"u + v = {addition}")
print(f"u - v = {subtraction}")
print(f"{scalar_constant}u = {scalar_multiplication}")
print(f"u . v = {dot_product}")
```

```python
# 3. Visualization

# Create the plot
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='red', label='u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v')
plt.quiver(0, 0, addition[0], addition[1], angles='xy', scale_units='xy', scale=1, color='green', label='u + v')
plt.quiver(0, 0, subtraction[0], subtraction[1], angles='xy', scale_units='xy', scale=1, color='magenta', label='u - v')
plt.quiver(0, 0, scalar_multiplication[0], scalar_multiplication[1], angles='xy', scale_units='xy', scale=1, color='yellow', label=f"{scalar_constant}u")

# Annotate the origin
plt.annotate("Origin\n[0, 0]", (0, 0), textcoords="offset points", xytext=(0, -30), ha='center')

# Set axis limits
lowestX, highestX = min(u[0], v[0], addition[0], subtraction[0], scalar_multiplication[0]), max(u[0], v[0], addition[0], subtraction[0], scalar_multiplication[0])
lowestY, highestY = min(u[1], v[1], addition[1], subtraction[1], scalar_multiplication[1]), max(u[1], v[1], addition[1], subtraction[1], scalar_multiplication[1])
lowest, highest = min(lowestX, lowestY), max(highestX, highestY)

plt.xlim(lowest - int(lowest / 3) - 1, highest + int(highest / 3) + 1)
plt.ylim(lowest - int(lowest / 3) - 1, highest + int(highest / 3) + 1)
plt.grid()

# Title and labels
plt.legend()
plt.title("Dynamic Vector Visualization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Save the plot (Must be called before showing the plot)
save_path = "vector_visualization.png"
plt.savefig(save_path)
print(f"Plot saved as '{save_path}'.")

# Show the plot
plt.show()
```
